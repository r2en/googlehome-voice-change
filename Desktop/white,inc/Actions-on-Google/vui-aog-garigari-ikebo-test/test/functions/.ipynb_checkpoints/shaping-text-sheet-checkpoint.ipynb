```python
import csv
```

```python
f = open('out-sheet.csv', 'r')
```

```python
reader = csv.reader(f)
```

```python
#含まれる改行を削除できるか？
#enumurate
for i, row in enumerate(reader):
    #print(row[1].replace('\n', ''))
    print('    ' + 'msg_{0}'.format(i) + ': ' + "['" + row[1].replace('\n', '') + "'], #" + row[0] + '\n') 
```

```python
with open('shaping-text-sheet.js', mode='w') as f:
    for row in reader:
        f.write('    ' + row[0] + ': ' + "['" + row[1] + "']," + '\n') 
        # voice : f.write('    ' + row[0] + ': ' + "['" + "']," + '\n') 
```
